## Abstract

This study develops a quantitative foreign exchange (FX) trading strategy that integrates carry trade with macroeconomic uncertainty analysis to optimize trade execution and risk management. The strategy is applied to six highly liquid currency pairs—CHF/HKD, CHF/EUR, CHF/JPY, CHF/GBP, CHF/CAD, and CHF/USD—selected based on their market depth, trading volume, and exposure to macroeconomic fluctuations.

Our approach is grounded in the traditional carry trade framework, where the Swiss franc (CHF) serves as the funding currency due to its persistently low interest rates. The strategy involves taking long positions in higher-yielding currencies while shorting CHF, capturing interest rate differentials as a primary source of returns. However, as carry trade performance is inherently sensitive to macroeconomic conditions, risk-off events, and shifts in monetary policy, we incorporate a macro uncertainty index to systematically assess external risk factors that impact FX markets.

Instead of relying on isolated financial news sentiment, we develop a comprehensive macro uncertainty measure, integrating economic indicators, central bank policies, geopolitical developments, and financial market volatility. This index enhances our ability to quantify systemic risks, refine position sizing, and adapt trade execution to prevailing macroeconomic conditions.

The strategy is empirically tested using daily FX data, evaluating its profitability, volatility-adjusted performance (e.g., Sharpe ratio), and sensitivity to macroeconomic risk factors. Our findings contribute to the broader field of quantitative FX trading by demonstrating the impact of macroeconomic uncertainty on carry trade efficiency, risk mitigation, and capital allocation optimization.

### Introduction

Foreign exchange (FX) markets are among the most actively traded and liquid financial markets globally, with a daily turnover exceeding $7.5 trillion (Bank for International Settlements, BIS). Given their deep liquidity, efficiency, and responsiveness to macroeconomic developments, FX markets present opportunities for systematic trading strategies that capitalize on interest rate differentials, market sentiment, and macroeconomic conditions. This study develops a quantitative carry trade strategy that incorporates macro uncertainty analysis to optimize trade execution and risk management.


#### 1.1. CHF as the Funding Currency
The Swiss franc (CHF) has long been considered a safe-haven currency, primarily due to Switzerland’s low inflation, stable political environment, and historically low interest rates. The Swiss National Bank (SNB) maintains one of the lowest policy rates among major central banks, making CHF a favorable funding currency for carry trade strategies.

Unlike traditional carry trade frameworks that rely on the Japanese yen (JPY) or U.S. dollar (USD) as funding currencies, our strategy shorts CHF to finance long positions in higher-yielding currencies. The rationale behind using CHF as the funding base is as follows:

Consistently Low Interest Rates: The SNB has maintained negative or near-zero interest rates for extended periods, allowing for cost-effective borrowing in CHF.

Safe-Haven Characteristics: CHF appreciates during risk-off periods, making it an ideal benchmark for assessing macroeconomic risk exposure in FX carry trades.
Policy Intervention and Market Impact: The SNB actively intervenes in FX markets to prevent excessive CHF appreciation, leading to short-term dislocations that may offer trading opportunities.

Diversification from JPY-Based Carry Trades: While JPY has historically been a funding currency, CHF-based funding provides an alternative perspective, particularly in European and North American macroeconomic contexts.

#### 1.2 Currency Pair Selection
To ensure optimal liquidity, macroeconomic exposure, and interest rate differentials, our strategy focuses on six major currency pairs involving the Swiss franc (CHF):

•	CHF/HKD: The Hong Kong dollar is managed under a pegged exchange rate regime with the U.S. dollar, creating opportunities for arbitrage and policy-driven mispricing. The Hong Kong Monetary Authority (HKMA) periodically intervenes in the FX market, influencing short-term fluctuations.

•	CHF/EUR: The euro is the second most traded currency worldwide and represents a major reserve currency. European Central Bank (ECB) policies, including interest rate adjustments and quantitative easing, significantly impact EUR valuation.

•	CHF/JPY: The Japanese yen has historically been used in carry trade strategies, as Japan's ultra-low interest rates make JPY a key funding currency. However, in this study, we shift the funding base to CHF while assessing JPY's sensitivity to risk sentiment, global trade flows, and Bank of Japan (BoJ) policies.

•	CHF/GBP: The British pound remains one of the most volatile major currencies due to Bank of England (BoE) monetary policy decisions, post-Brexit economic realignments, and inflationary pressures. GBP’s volatility enhances trading opportunities, particularly in carry trade risk management.

•	CHF/CAD: The Canadian dollar is highly correlated with crude oil prices, given Canada’s status as a major energy exporter. The Bank of Canada’s (BoC) policy stance, energy market trends, and trade relations with the U.S. make USD/CAD a strong candidate for macro-driven analysis.

•	CHF/USD: The U.S. dollar remains the world’s primary reserve currency, and the Federal Reserve's monetary policy decisions significantly influence global capital flows. USD movements often reflect risk-on and risk-off sentiment, making CHF/USD a key benchmark pair in evaluating macroeconomic risk.

#### 1.3 Historical Context of Carry Trade in Quantitative Trading
The carry trade strategy has long been a cornerstone of foreign exchange (FX) markets, with historical roots tracing back to the liberalization of capital markets and the emergence of global interest rate differentials. The strategy gained prominence in the late 20th century, particularly after the collapse of the Bretton Woods system in the early 1970s, when currencies shifted from fixed exchange rates to a floating regime. This transition introduced new profit-making opportunities by allowing traders to exploit systematic interest rate disparities across economies.

By the 1980s and 1990s, hedge funds and institutional investors increasingly adopted quantitative models to optimize carry trade execution, integrating macroeconomic indicators and time-series analysis to enhance risk-adjusted returns. The Japanese yen (JPY), characterized by persistently low interest rates due to the Bank of Japan’s (BoJ) accommodative monetary policies, became the primary funding currency, while high-yielding currencies such as the Australian dollar (AUD), New Zealand dollar (NZD), and emerging market currencies were preferred for long positions.

Despite its historical profitability, carry trade strategies exhibit distinct cyclical behavior, thriving during periods of low volatility and stable monetary policy but suffering sharp reversals during financial crises and risk-off events. The 2008 Global Financial Crisis (GFC) demonstrated the vulnerabilities of carry trades, as widespread deleveraging and capital flight led to rapid unwinding of positions, resulting in severe FX market dislocations. This highlighted the importance of risk management and adaptive models in carry trade strategies.

#### 1.4 Trading Strategy and Risk Management
Foreign exchange (FX) carry trade strategies are fundamentally designed to exploit interest rate differentials by borrowing in low-yielding currencies and reallocating capital into higher-yielding counterparts. While historically effective, conventional carry trade models are highly susceptible to macroeconomic shifts, sudden volatility spikes, and global risk sentiment fluctuations. To enhance robustness and risk-adjusted returns, this study integrates a systematic risk management approach that incorporates macro uncertainty analysis and dynamic hedging mechanisms, ensuring adaptability to evolving market conditions.

The core strategy is structured around leveraging CHF as a funding currency, given its historically low interest rates, and taking long positions in selected higher-yielding currencies. However, the success of a carry trade depends not only on yield differentials but also on exchange rate stability and macroeconomic conditions. Unanticipated currency depreciations, liquidity constraints, and risk-off sentiment can severely undermine returns, making risk management a critical component of execution. Instead of adopting a purely static approach, this strategy dynamically adjusts exposure based on volatility assessments and external macroeconomic risks.

To quantify external risks that influence FX price movements, this study develops a macro uncertainty index, incorporating economic indicators, central bank policies, geopolitical events, and financial market instability. This index, derived from Bloomberg data, serves as a primary tool for evaluating systemic risks and determining optimal trade sizing and hedge adjustments. By systematically integrating macroeconomic risk factors, the strategy ensures that carry trade positions are adjusted to reflect prevailing economic conditions, mitigating downside risk while maximizing return potential.

Furthermore, historical volatility calculations using daily FX data provide additional risk mitigation by informing position sizing and leverage decisions. By continuously monitoring volatility trends and macroeconomic uncertainty, the strategy adapts its exposure to ensure resilience against adverse market shifts. The integration of these risk management components enhances the overall performance of the carry trade model, creating a framework that balances profitability with systematic risk mitigation.

This approach represents an advancement over traditional carry trade methodologies by aligning macro-driven risk assessment with FX market dynamics, enabling a more resilient and data-driven decision-making process. The following sections will outline the empirical methodology used to evaluate the performance and stability of this strategy under varying macroeconomic conditions.

```python
def plot_yield(yield_df, t_saron, country, yield_chf):

    # Rename columns for consistency
    # t_saron.rename(columns={'Date': 'Dates', 'Mid Price': 'SARON'}, inplace=True)

    # Convert Dates to datetime format
    t_saron = t_saron[['Dates', 'SARON close of trading']]
    t_saron['Dates'] = pd.to_datetime(t_saron['Dates'])
    yield_df['Dates'] = pd.to_datetime(yield_df['Dates'])
    yield_chf['Dates'] = pd.to_datetime(yield_chf['Dates'])

    # Merge the datasets on Dates to align SARON with the bond yields
    yield_df_merged = pd.merge(yield_df, t_saron, on='Dates', how='inner')

    # Plot YLD_YTM_MID for 1Y, 5Y, and 10Y bonds along with SARON
    plt.figure(figsize=(12, 6))
    plt.plot(yield_df_merged['Dates'], yield_df_merged['YLD_YTM_MID_1Y'], label=f'{country} 1Y Yield', linestyle='-')
    plt.plot(yield_df_merged['Dates'], yield_df_merged['YLD_YTM_MID_5Y'], label=f'{country} 5Y Yield', linestyle='-')
    if 'YLD_YTM_MID_10Y' in yield_df.columns:
        plt.plot(yield_df_merged['Dates'], yield_df_merged['YLD_YTM_MID_10Y'], label=f'{country}10Y Yield', linestyle='-')
    plt.plot(yield_df_merged['Dates'], yield_df_merged['SARON close of trading'], label='SARON close of trading', linestyle='--', color='grey')

    # Plot YLD_YTM_MID for 1Y, 5Y, and 10Y bonds along with SARON
    # plt.figure(figsize=(12, 6))
    plt.plot(yield_chf['Dates'], yield_chf['YLD_YTM_MID_1Y'], label='CHF 1Y Yield', linestyle='-')
    plt.plot(yield_chf['Dates'], yield_chf['YLD_YTM_MID_5Y'], label='CHF 5Y Yield', linestyle='-')
    plt.plot(yield_chf['Dates'], yield_chf['YLD_YTM_MID_10Y'], label='CHF 10Y Yield', linestyle='-')


    plt.xlabel('Date')
    plt.ylabel('Rate (%)')
    plt.title(f'{country} Yield to Maturity (Mid) for Bonds and SARON Rate')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def plot_px_spread(df, country):
    # Forward fill missing data
    df.ffill(inplace=True)

    # Plot Bid-Ask spread as shaded area for 1Y, 5Y, and 10Y bond prices
    if 'PX_BID_10Y' in df.columns:
        fig, axes = plt.subplots(3, 1, figsize=(12, 12), sharex=True)
    else:
        fig, axes = plt.subplots(2, 1, figsize=(12, 12), sharex=True)



    # 1Y bond
    axes[0].fill_between(df['Dates'], df['PX_BID_1Y'], df['PX_ASK_1Y'], color='blue', alpha=0.3, label='Bid-Ask Spread')
    axes[0].plot(df['Dates'], df['PX_MID_1Y'], label='PX_MID_1Y', linestyle='-', color='yellow')
    axes[0].set_ylabel('Price')
    axes[0].set_title(f'{country} 1Y Bond Prices (Bid-Ask Spread and Mid)')
    axes[0].legend()
    axes[0].grid()

    # 5Y bond
    axes[1].fill_between(df['Dates'], df['PX_BID_5Y'], df['PX_ASK_5Y'], color='blue', alpha=0.3, label='Bid-Ask Spread')
    axes[1].plot(df['Dates'], df['PX_MID_5Y'], label='PX_MID_5Y', linestyle='-', color='yellow')
    axes[1].set_ylabel('Price')
    axes[1].set_title(f'{country} 5Y Bond Prices (Bid-Ask Spread and Mid)')
    axes[1].legend()
    axes[1].grid()

    # 10Y bond
    if 'PX_BID_10Y' in df.columns:
        axes[2].fill_between(df['Dates'], df['PX_BID_10Y'], df['PX_ASK_10Y'], color='blue', alpha=0.3, label='Bid-Ask Spread')
        axes[2].plot(df['Dates'], df['PX_MID_10Y'], label='PX_MID_10Y', linestyle='-', color='yellow')
        axes[2].set_ylabel('Price')
        axes[2].set_title(f'{country} 10Y Bond Prices (Bid-Ask Spread and Mid)')
        axes[2].set_xlabel('Date')
        axes[2].legend()
        axes[2].grid()

    plt.tight_layout()
    plt.show()
```

```python
close_prices = pd.DataFrame()

for pair, data in fx_data.items():
    close_col = [col for col in data.columns if 'close' in col.lower()][0]
    close_prices[pair] = data[close_col]

# Make sure dates are sorted in ascending order
close_prices.sort_index(inplace=True)

# Calculate daily returns
daily_returns = close_prices.pct_change().dropna()

# Calculate rolling volatility (30-day standard deviation of returns, annualized)
rolling_volatility = daily_returns.rolling(window=30).std() * np.sqrt(252) * 100  # Annualized and in percentage

# Calculate rolling correlation (90-day) with USD as the base
if 'CHF/USD' in daily_returns.columns:
    usd_returns = daily_returns['CHF/USD']
    rolling_correlations = pd.DataFrame()

    for currency in daily_returns.columns:
        if currency != 'CHF/USD':
            rolling_correlations[f'{currency} vs USD'] = daily_returns[currency].rolling(90).corr(usd_returns)

# 1. Plot closing prices
plt.figure(figsize=(14, 8))
for pair in close_prices.columns:
    plt.plot(close_prices.index, close_prices[pair], label=pair)

plt.title('CHF Exchange Rates (Daily Closing Prices)')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('chf_exchange_rates.png')
plt.show()

# 2. Plot daily returns
plt.figure(figsize=(14, 8))
for pair in daily_returns.columns:
    plt.plot(daily_returns.index, daily_returns[pair] * 100, label=pair, alpha=0.7)  # Convert to percentage

plt.title('Daily Returns of CHF Exchange Rates (%)')
plt.xlabel('Date')
plt.ylabel('Daily Return (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('chf_daily_returns.png')
plt.show()

# 3. Plot rolling volatility
plt.figure(figsize=(14, 8))
for pair in rolling_volatility.columns:
    plt.plot(rolling_volatility.index, rolling_volatility[pair], label=pair)

plt.title('30-Day Rolling Volatility of CHF Exchange Rates (Annualized %)')
plt.xlabel('Date')
plt.ylabel('Volatility (%)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('chf_volatility.png')
plt.show()

# 4. Plot rolling correlations
if 'rolling_correlations' in locals() and not rolling_correlations.empty:
    plt.figure(figsize=(14, 8))
    for col in rolling_correlations.columns:
        plt.plot(rolling_correlations.index, rolling_correlations[col], label=col)

    plt.title('90-Day Rolling Correlations with CHF/USD')
    plt.xlabel('Date')
    plt.ylabel('Correlation Coefficient')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('chf_correlations.png')
    plt.show()

# 5. Correlation heatmap (using the full dataset)
plt.figure(figsize=(10, 8))
correlation_matrix = daily_returns.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Correlation Matrix of CHF Exchange Rate Returns')
plt.tight_layout()
plt.savefig('chf_correlation_heatmap.png')
plt.show()

# 6. Risk-Return plot (for carry trade analysis)
# Calculate annualized returns and risks
annualized_returns = daily_returns.mean() * 252 * 100  # Annualized and in percentage
annualized_risk = daily_returns.std() * np.sqrt(252) * 100  # Annualized and in percentage

plt.figure(figsize=(10, 8))
plt.scatter(annualized_risk, annualized_returns, s=50)

# Label each point
for i, pair in enumerate(daily_returns.columns):
    plt.annotate(pair,
                 (annualized_risk[i], annualized_returns[i]),
                 textcoords="offset points",
                 xytext=(0,10),
                 ha='center')

plt.axhline(y=0, color='r', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='r', linestyle='-', alpha=0.3)
plt.title('Risk-Return Profile of CHF Exchange Rates')
plt.xlabel('Annualized Risk (%)')
plt.ylabel('Annualized Return (%)')
plt.grid(True)
plt.tight_layout()
plt.savefig('chf_risk_return.png')
plt.show()

# 7. Calculate and show potential carry trade returns (without interest rates)
print("\nRisk-Return Analysis (Without Interest Rate Differential):")
for pair in daily_returns.columns:
    print(f"{pair}:")
    print(f"  Annual Return: {annualized_returns[pair]:.2f}%")
    print(f"  Annual Volatility: {annualized_risk[pair]:.2f}%")
    print(f"  Sharpe Ratio (assuming 0% risk-free rate): {annualized_returns[pair]/annualized_risk[pair]:.3f}")
    print()
```

Based on the provided exchange rate data and charts:

1. Volatility Analysis

  The 30-day rolling volatility chart (Image 1) reveals several important patterns:

    - CHF/JPY consistently exhibits the highest volatility, fluctuating between 8-10% annualized. This level of volatility makes it less suitable for risk-averse carry trade strategies, as large fluctuations could offset interest rate gains.

    - CHF/EUR demonstrates the lowest volatility (3-4% annualized), highlighting a relatively stable exchange rate between these two European currencies. This stability makes CHF/EUR an attractive pair for carry trades seeking lower risk.

    - CHF/USD and CHF/HKD exhibit similar volatility levels (~6-7%), likely due to HKD's peg to USD. This implies that any shifts in USD rates or monetary policy could impact CHF/HKD in a highly correlated manner.

    - CHF/GBP has shown a gradual decline in volatility, decreasing from ~6.5% to ~5% over the observed period. This trend suggests that CHF/GBP is becoming a more predictable currency pair for traders.

  Given this volatility structure, EUR-funded or EUR-targeted carry trades involving CHF would provide the most stable basis for carry trade strategies. CHF/JPY, due to its high volatility, may be riskier for such trades.

***Strategic Carry Trade Implications***

1. CHF as a Funding Currency:

  The low yield environment in CHF-denominated bonds and the persistently low SARON rate make the Swiss Franc an attractive funding currency for carry trade strategies. Borrowing CHF incurs minimal interest costs, increasing the appeal of funding in CHF and investing in higher-yielding assets.

  However, the strengthening of CHF against most major currencies, as evident from the negative annualized returns in the charts, poses a significant currency risk. A strengthening CHF could erode the gains from the interest rate differential, making unhedged carry trades less profitable or even loss-making over time.


2. Optimal Currency Pairs:

  Selecting the right currency pairs is crucial in optimizing risk-adjusted returns when using CHF as a funding currency.

  - CHF-EUR:

    Offers the lowest volatility, making it the most stable choice for a CHF carry trade. However, due to the modest yield differentials and limited exchange rate fluctuations, the return potential is relatively low. Works well for low-risk, long-term carry strategies where stability is prioritized over high returns.

  - CHF-JPY:

    Provides significant diversification benefits, as CHF/JPY has a low correlation with other CHF pairs (seen in correlation analysis). However, substantial volatility (8-10%) introduces exchange rate risk, which could wipe out carry trade gains. This pair is better suited for traders willing to accept higher risk in exchange for greater return potential.

  - Avoiding CHF-HKD Trades:

    Due to HKD’s peg to USD, CHF-HKD movements are almost identical to CHF-USD. The high correlation (~0.99) between CHF/USD and CHF/HKD suggests no diversification benefit from holding both. Unless there is an explicit trading advantage, CHF-HKD carry trades are redundant compared to CHF-USD trades.

3. Hedging by futures:

  Assumption: Foreign Exchange Futures serve as a hedging instrument to mitigate potential losses caused by exchange rate fluctuations.

  Scenario: Imagine borrowing 100 Swiss Francs (CHF) from Switzerland and converting them into Japanese Yen (JPY) at an exchange rate of 166.61, resulting in 16,661 JPY. After a month, you convert the yen back to CHF. If during this period the CHF/JPY rate increases, you incur a loss; if it decreases, you gain a profit. Essentially, you profit when the Japanese Yen strengthens and the Swiss Franc weakens, which causes the CHF/JPY rate to drop. This position is akin to holding a long position in JPY or a short position in CHF (that is, shorting the CHF/JPY rate or going long on the JPY/CHF rate).

  Intuition: This strategy is similar to how an oil company hedges against falling oil prices by shorting oil futures to offset potential losses on its oil stock.

  Potential Drawback:
  1. Sometimes we don't need to hedge to potential loss, since the exchange rate's change is too small and we would pay more transaction fees. Therefore, we could adjust the hedging position by forecasting volatility (as in point 4).

  2. Futures orders are based on different expiration date, and we are using generic price data, we would face more challenges in reality. And the generic futures price data from Bloomberg have some date with None data, this would shrink our backtest data.

4. Risk Management Considerations:

  The forex carry trade relies on profiting from the interest rate differentials between currencies. However, because interest rates are heavily influenced by political and macroeconomic factors, carry trades can be unexpectedly risky. To manage this risk, our strategy incorporates hedging with futures alongside a risk threshold based on a macroeconomic indicator.

  We propose using the Economic Policy Uncertainty (EPU) Index—a Bloomberg-generated measure of policy-related uncertainty derived from major press from respective countres. This index filters articles for terms related to policy uncertainty as well as macro factors related to tax, economic forecasts etc. By monitoring the index, traders gain more information about potential volatility range, and compare the value with different threshold based on users' risk apetite. With this, we can proactively make decision on when to hedge the carry trade with futures contract to navigate sudden market shifts more effectively and tailor risk management to individual investor needs.





# **Conclusion**

The Swiss Franc demonstrates characteristics of an attractive funding currency due to Switzerland's traditionally low interest rate environment. However, its tendency to strengthen against most major currencies (particularly JPY, USD, and HKD) creates a counterbalancing risk that could erode interest rate advantages in carry trades.
For currency-only carry trades (without considering interest differentials), the data suggests CHF/EUR as the most stable foundation, while CHF/JPY offers diversification but with substantially higher volatility. A strategically weighted basket approach could potentially balance these characteristics for optimal risk-adjusted returns.
