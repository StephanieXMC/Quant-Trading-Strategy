This study develops a quantitative foreign exchange (FX) trading strategy that integrates carry trade with macroeconomic uncertainty analysis to optimize trade execution and risk management. ### Introduction

Foreign exchange (FX) markets are among the most actively traded and liquid financial markets globally, with a daily turnover exceeding $7.5 trillion (Bank for International Settlements, BIS). Given their deep liquidity, efficiency, and responsiveness to macroeconomic developments, FX markets present opportunities for systematic trading strategies that capitalize on interest rate differentials, market sentiment, and macroeconomic conditions. This study develops a quantitative carry trade strategy that incorporates macro uncertainty analysis to optimize trade execution and risk management.


#### 1.1. CHF as the Funding Currency
The Swiss franc (CHF) has long been considered a safe-haven currency, primarily due to Switzerland’s low inflation, stable political environment, and historically low interest rates. The Swiss National Bank (SNB) maintains one of the lowest policy rates among major central banks, making CHF a favorable funding currency for carry trade strategies.

Unlike traditional carry trade frameworks that rely on the Japanese yen (JPY) or U.S. dollar (USD) as funding currencies, our strategy shorts CHF to finance long positions in higher-yielding currencies. The rationale behind using CHF as the funding base is as follows:

Consistently Low Interest Rates: The SNB has maintained negative or near-zero interest rates for extended periods, allowing for cost-effective borrowing in CHF.

Safe-Haven Characteristics: CHF appreciates during risk-off periods, making it an ideal benchmark for assessing macroeconomic risk exposure in FX carry trades.
Policy Intervention and Market Impact: The SNB actively intervenes in FX markets to prevent excessive CHF appreciation, leading to short-term dislocations that may offer trading opportunities.

Diversification from JPY-Based Carry Trades: While JPY has historically been a funding currency, CHF-based funding provides an alternative perspective, particularly in European and North American macroeconomic contexts.

#### 1.3 Historical Context of Carry Trade in Quantitative Trading
The carry trade strategy has long been a cornerstone of foreign exchange (FX) markets, with historical roots tracing back to the liberalization of capital markets and the emergence of global interest rate differentials. The strategy gained prominence in the late 20th century, particularly after the collapse of the Bretton Woods system in the early 1970s, when currencies shifted from fixed exchange rates to a floating regime. This transition introduced new profit-making opportunities by allowing traders to exploit systematic interest rate disparities across economies.

By the 1980s and 1990s, hedge funds and institutional investors increasingly adopted quantitative models to optimize carry trade execution, integrating macroeconomic indicators and time-series analysis to enhance risk-adjusted returns. The Japanese yen (JPY), characterized by persistently low interest rates due to the Bank of Japan’s (BoJ) accommodative monetary policies, became the primary funding currency, while high-yielding currencies such as the Australian dollar (AUD), New Zealand dollar (NZD), and emerging market currencies were preferred for long positions.

Despite its historical profitability, carry trade strategies exhibit distinct cyclical behavior, thriving during periods of low volatility and stable monetary policy but suffering sharp reversals during financial crises and risk-off events. The 2008 Global Financial Crisis (GFC) demonstrated the vulnerabilities of carry trades, as widespread deleveraging and capital flight led to rapid unwinding of positions, resulting in severe FX market dislocations. This highlighted the importance of risk management and adaptive models in carry trade strategies.

###1.5 Hedging in Carry Trade Strategies

While carry trade strategies offer attractive returns by exploiting interest rate differentials, they are inherently exposed to foreign exchange (FX) risk. The profitability of a carry trade depends not only on yield spreads but also on stable or favorable exchange rate movements. However, FX markets are highly volatile, influenced by macroeconomic uncertainty, monetary policy shifts, and sudden risk-off events, which can lead to significant losses for unhedged carry trade positions.

To mitigate these risks, traders employ hedging techniques that protect against adverse currency fluctuations while preserving potential returns. Two primary approaches are commonly used: forward hedging and option-based hedging. Forward contracts lock in future exchange rates, eliminating FX uncertainty but also restricting upside gains from favorable rate movements. In contrast, options provide asymmetric protection, allowing traders to limit downside risk while retaining the ability to benefit from positive FX shifts.

This study evaluates the effectiveness of both forward contracts (Section 8) and USD/CHF options (Section 9) as hedging mechanisms within a CHF-funded carry trade framework. We empirically compare their impact on risk-adjusted returns and assess their trade-offs in managing currency exposure. By integrating these hedging strategies, we aim to enhance the stability and robustness of carry trade positions in dynamic FX market conditions.

### **Currency Data Analysis and Plot**

```python
close_prices = pd.DataFrame()

for pair, data in fx_data.items():
    close_col = [col for col in data.columns if 'close' in col.lower()][0]
    close_prices[pair] = data[close_col]

# Make sure dates are sorted in ascending order
close_prices.sort_index(inplace=True)

# Calculate daily returns
daily_returns = close_prices.pct_change().dropna()

# Calculate rolling volatility (30-day standard deviation of returns, annualized)
rolling_volatility = daily_returns.rolling(window=30).std() * np.sqrt(252) * 100  # Annualized and in percentage

# Calculate rolling correlation (90-day) with USD as the base
if 'CHF_USD' in daily_returns.columns:
    usd_returns = daily_returns['CHF_USD']
    rolling_correlations = pd.DataFrame()

    for currency in daily_returns.columns:
        if currency != 'CHF_USD':
            rolling_correlations[f'{currency} vs USD'] = daily_returns[currency].rolling(90).corr(usd_returns)

# 1. Plot closing prices
plt.figure(figsize=(14, 8))
for pair in close_prices.columns:
    plt.plot(close_prices.index, close_prices[pair], label=pair)

plt.title('CHF Exchange Rates (Daily Closing Prices)')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('chf_exchange_rates.png')
plt.show()
```

```python
# Define the updated function to plot yield data
def plot_yield(yield_df, t_saron, country, yield_chf, ax):
    """Plots yield curves for 1Y, 5Y, and 10Y maturities along with SARON and CHF yields."""

    # Ensure 'Dates' is in datetime format
    t_saron['Dates'] = pd.to_datetime(t_saron['Dates'])
    yield_df['Dates'] = pd.to_datetime(yield_df['Dates'])
    yield_chf['Dates'] = pd.to_datetime(yield_chf['Dates'])

    # Merge the datasets on Dates to align SARON with bond yields
    yield_df_merged = pd.merge(yield_df, t_saron[['Dates', 'SARON close of trading']], on='Dates', how='inner')

    # Plot YLD_YTM_MID for 1Y, 5Y, and 10Y bonds along with SARON
    ax.plot(yield_df_merged['Dates'], yield_df_merged['YLD_YTM_MID_1Y'], label=f'{country} 1Y Yield', linestyle='-')
    ax.plot(yield_df_merged['Dates'], yield_df_merged['YLD_YTM_MID_5Y'], label=f'{country} 5Y Yield', linestyle='-')

    if 'YLD_YTM_MID_10Y' in yield_df.columns:
        ax.plot(yield_df_merged['Dates'], yield_df_merged['YLD_YTM_MID_10Y'], label=f'{country} 10Y Yield', linestyle='-')

    ax.plot(yield_df_merged['Dates'], yield_df_merged['SARON close of trading'], label='SARON close of trading', linestyle='--', color='grey')

    # Plot CHF yields for comparison
    ax.plot(yield_chf['Dates'], yield_chf['YLD_YTM_MID_1Y'], label='CHF 1Y Yield', linestyle='-')
    ax.plot(yield_chf['Dates'], yield_chf['YLD_YTM_MID_5Y'], label='CHF 5Y Yield', linestyle='-')
    ax.plot(yield_chf['Dates'], yield_chf['YLD_YTM_MID_10Y'], label='CHF 10Y Yield', linestyle='-')

    ax.set_xlabel('Date')
    ax.set_ylabel('Rate (%)')
    ax.set_title(f'{country} Yield to Maturity (Mid) for Bonds and SARON Rate')
    ax.legend()
    ax.grid()

# List of yield data and corresponding currency labels
yield_data_list = [
    (t_yield_gbp, "GBP"),
    (t_yield_jpy, "JPY"),
    # (t_yield_eur, "EUR"),
    (t_yield_cad, "CAD"),
    (t_yield_usd, "USD"),
    (t_yield_hkd, "HKD"),
    (t_yield_dem, "DEM"),
    # (t_yield_chf, "CHF")
]

# Create a grid for plots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 11))  # 4 rows, 2 columns

# Flatten axes array for easy iteration
axes = axes.flatten()

# Plot each yield dataset
for i, (yield_data, label) in enumerate(yield_data_list):
    plot_yield(yield_data, t_saron, label, t_yield_chf, axes[i])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
def compute_zcb_curve(spot_rates_curve):
    """
    Computes the zero-coupon bond (ZCB) yield curve from a given spot rates curve.

    Parameters:
    -----------
    spot_rates_curve : pandas.DataFrame
        A DataFrame where each column represents a different spot rate curve,
        and the index contains the tenors (in years).

    Returns:
    --------
    pandas.DataFrame
        A DataFrame of zero-coupon bond (ZCB) rates with the same structure as `spot_rates_curve`.

    Method:
    -------
    - For each tenor greater than 0.001 years, the function estimates the zero-coupon rate.
    - It interpolates previously computed ZCB rates for discounting coupon payments.
    - The formula used accounts for preceding coupon values and derives the ZCB rate
      using the logarithmic discounting formula.

    Notes:
    ------
    - The function assumes **quarterly coupon payments** (every 3 months).
    - Linear interpolation is used for discounting preceding coupon payments.
    - The calculations use continuous compounding.

    """
    zcb_rates = spot_rates_curve.copy()
    for curve in spot_rates_curve.columns:
        spot = spot_rates_curve[curve]
        for tenor, spot_rate in spot.items():
            if tenor>0.001:
                times = np.arange(tenor - 0.25, 0, step=-0.25)[::-1]
                coupon_quarterly = 0.25 * spot_rate
                z = np.interp(times, zcb_rates[curve].index.values, zcb_rates[curve].values) # Linear interpolation
                preceding_coupons_val = (coupon_quarterly*np.exp(-z*times)).sum()
                zcb_rates[curve][tenor] = -np.log((1-preceding_coupons_val)/(1+coupon_quarterly))/tenor


    return zcb_rates
```

```python
def get_wednesdays_with_adjustments(start_date, end_date, list1):
    """
    Finds all Wednesday dates within a given date range. If a Wednesday is missing
    in list1, it is replaced with the next available date in list1. Duplicates are removed.

    Parameters:
    -----------
    start_date : str or datetime
        The start of the date range (format: 'YYYY-MM-DD').
    end_date : str or datetime
        The end of the date range (format: 'YYYY-MM-DD').
    list1 : list of pandas.Timestamp
        A list of valid dates.

    Returns:
    --------
    pandas.DatetimeIndex
        A sorted, unique list of adjusted Wednesday dates.
    """
    # Generate all Wednesday dates within the range
    all_dates = pd.date_range(start=start_date, end=end_date, freq='D')
    wednesdays = all_dates[all_dates.weekday == 2]  # Wednesday is weekday 2

    # Convert list1 to a sorted set for fast lookup & searching
    valid_dates = pd.to_datetime(sorted(set(list1)))

    # Adjust Wednesdays if they are missing from list1
    adjusted_wednesdays = set()  # Using a set to avoid duplicates
    for wednesday in wednesdays:
        if wednesday not in valid_dates:
            # Find the next available valid date
            next_valid_date = next((date for date in valid_dates if date > wednesday), None)
            if next_valid_date:
                adjusted_wednesdays.add(next_valid_date)
        else:
            adjusted_wednesdays.add(wednesday)

    # Return a sorted, unique list of dates
    return pd.DatetimeIndex(sorted(adjusted_wednesdays))



gbp_wkly_dates = get_wednesdays_with_adjustments(start_date, end_date, pd.to_datetime(gbp_dates))
jpy_wkly_dates = get_wednesdays_with_adjustments(start_date, end_date, pd.to_datetime(jpy_dates))
usd_wkly_dates = get_wednesdays_with_adjustments(start_date, end_date, pd.to_datetime(usd_dates))
hkd_wkly_dates = get_wednesdays_with_adjustments(start_date, end_date, pd.to_datetime(hkd_dates))
dem_wkly_dates = get_wednesdays_with_adjustments(start_date, end_date, pd.to_datetime(dem_dates))
cad_wkly_dates = get_wednesdays_with_adjustments(start_date, end_date, pd.to_datetime(cad_dates))
chf_wkly_dates = get_wednesdays_with_adjustments(start_date, end_date, pd.to_datetime(chf_dates))

```

```python
def bond_price(zcb, coupon_rate, tenor):
    """
    Calculate the price of a bond using a given zero-coupon bond (ZCB) curve.

    Parameters:
    -----------
    zcb : pandas.Series
        A zero-coupon bond yield curve, where the index represents maturities
        (in years) and the values represent continuously compounded yields.
    coupon_rate : float
        The annualized coupon rate of the bond (as a decimal).
    tenor : float
        The bond's maturity in years.

    Returns:
    --------
    float
        The present value (price) of the bond.

    Notes:
    ------
    - The bond pays semi-annual coupons, as indicated by the step size of 0.5 in the `times` array.
    - The function performs linear interpolation on the ZCB curve to estimate rates at required time points.
    - The bond price consists of two components:
      1. The present value of the face value discounted at the interpolated rate.
      2. The present value of semi-annual coupon payments discounted at the interpolated rates.
    """
    times = np.arange(tenor, 0, step=-0.25)[::-1]
    if times.shape[0]==0:
        p = 1.0
    else:
        r = np.interp(times, zcb.index.values, zcb.values) # Linear interpolation
        p = np.exp(-tenor*r[-1]) + 0.25 * coupon_rate * np.exp(-r*times).sum()
    return p
```

```python
def carry_trade_strat(t_currency_rate, OIS_data, country_yields, zeros_data,
                      lending_currency, start_date, end_date, notional=10_000_000,
                        leverage_ratio=5):
    """
    Implements a carry trade strategy for a given country and currency.

    Parameters:
    -----------
    t_currency_rate : pandas.DataFrame
        A DataFrame containing historical foreign exchange rates.
    OIS_data : pandas.DataFrame
        A DataFrame containing historical OIS rates.
    country_yields : pandas.DataFrame
        A DataFrame containing historical yields for the given country.
    zeros_data : pandas.DataFrame
        A DataFrame containing zero-coupon bond rates for the given country.
    lending_currency : str
        The currency in which the investor will lend.
    start_date : str or datetime
        The start date of the analysis (format: 'YYYY-MM-DD').
    end_date : str or datetime
        The end date of the analysis (format: 'YYYY-MM-DD').
    notional : float, optional (default=10_000_000)
        The initial notional amount to invest.
    leverage_ratio : float, optional (default=5)
        The leverage ratio for the carry trade strategy.

    Returns:
    --------

    float
        The cumulative profit or loss from the carry trade strategy.
    dict
        A dictionary of weekly profits or losses from the carry trade strategy.
    dict
        A dictionary of detailed profits or losses from the carry trade strategy.
    """
    zeros_data.columns = pd.to_datetime(zeros_data.columns)

    country_yields['tenor_mapped'] = country_yields['tenor_mapped'].astype(int)
    country_yields['date'] = pd.to_datetime(country_yields['date'])

    wkly_dates = get_wednesdays_with_adjustments(start_date, end_date, country_yields['date'].to_list())
    wkly_dates = wkly_dates.intersection(OIS_data['date'].to_list())
    wkly_dates = wkly_dates.intersection(zeros_data.columns.to_list())

    detailed_pnls = {}
    weekly_pnls = {}
    debt = notional * (1-(1/leverage_ratio))
    cumulative_pnl = 0

    for i, date in enumerate(wkly_dates):
        if i == len(wkly_dates) - 1:
            break
        if wkly_dates[i+1] - date > timedelta(days=8):
            continue


        row_index = OIS_data[OIS_data['date'] == date].index[0]
        OIS_rate = OIS_data.loc[row_index,'OIS']
        borrowing_rate = (OIS_rate + 0.005) / 100 / 52
        # borrowing_rate = (OIS_rate + 0.005) / 52

        # get lending rate
        lending_rate = country_yields[(country_yields['tenor_mapped'] == 5) & (country_yields['date'] == date)]['yield'].values[0]

        if lending_rate / 52  < borrowing_rate:
            continue
        else:
            #  t = 0
            initial_fx_rate_foreign_chf = t_currency_rate[(t_currency_rate['date'] == date) &
                                                  (t_currency_rate['code'] == lending_currency)]['rate'].values[0]
            initial_market_val = notional * initial_fx_rate_foreign_chf
            initial_price = bond_price(zeros_data[date], lending_rate, 5)
            qty_bonds = initial_market_val / initial_price

            #  t = 1
            next_date = wkly_dates[i+1]
            time_elapsed = (next_date - date).days / 365
            tenor = 5 - time_elapsed
            final_price = bond_price(zeros_data[next_date], lending_rate, tenor)
            final_market_val = final_price * qty_bonds
            bond_price_change = final_market_val - initial_market_val
            final_fx_rate_foreign_chf = t_currency_rate[(t_currency_rate['date'] == next_date) &
                                                (t_currency_rate['code'] == lending_currency)]['rate'].values[0]
            bond_price_change_chf = bond_price_change / final_fx_rate_foreign_chf

            #  t = 0
            initial_fx_rate_chf = t_currency_rate[(t_currency_rate['date'] == date) &
                                              (t_currency_rate['code'] == lending_currency)]['rate'].values[0]
            debt_lending = debt * initial_fx_rate_chf
            #  t = 1
            interest_payment = borrowing_rate * (debt_lending)
            final_fx_rate_chf = t_currency_rate[(t_currency_rate['date'] == next_date) &
                                            (t_currency_rate['code'] == lending_currency)]['rate'].values[0]
            interest_payment_chf = interest_payment / final_fx_rate_chf
            fx_pnl_chf = (debt * initial_fx_rate_chf / final_fx_rate_chf) - debt

            total_pnl = bond_price_change_chf - interest_payment_chf + fx_pnl_chf
            cumulative_pnl += total_pnl

            weekly_pnls[date] = total_pnl
            detailed_pnls[next_date] = {
                                'bond_price_change_usd': bond_price_change_chf,
                                'interest_payment_usd': interest_payment_chf,
                                'fx_pnl_usd': fx_pnl_chf,
                                'total_pnl': total_pnl}

    return cumulative_pnl, weekly_pnls, detailed_pnls

gbp_cum_pnl, gbp_wkly_pnls, gbp_detailed_pnls = carry_trade_strat(t_currency_rate, OIS_data, yield_gbp_filtered, gbp_zeros, 'CHF_GBP', start_date, end_date, notional=10_000_000, leverage_ratio=5)
jpy_cum_pnl, jpy_wkly_pnls, jpy_detailed_pnls = carry_trade_strat(t_currency_rate, OIS_data, yield_jpy_filtered, jpy_zeros, 'CHF_JPY', start_date, end_date, notional=10_000_000, leverage_ratio=5)
usd_cum_pnl, usd_wkly_pnls, usd_detailed_pnls = carry_trade_strat(t_currency_rate, OIS_data, yield_usd_filtered, usd_zeros, 'CHF_USD', start_date, end_date, notional=10_000_000, leverage_ratio=5)
hkd_cum_pnl, hkd_wkly_pnls, hkd_detailed_pnls = carry_trade_strat(t_currency_rate, OIS_data, yield_hkd_filtered, hkd_zeros, 'CHF_HKD', start_date, end_date, notional=10_000_000, leverage_ratio=5)
dem_cum_pnl, dem_wkly_pnls, dem_detailed_pnls = carry_trade_strat(t_currency_rate, OIS_data, yield_dem_filtered, dem_zeros, 'CHF_EUR', start_date, end_date, notional=10_000_000, leverage_ratio=5)
cad_cum_pnl, cad_wkly_pnls, cad_detailed_pnls = carry_trade_strat(t_currency_rate, OIS_data, yield_cad_filtered, cad_zeros, 'CHF_CAD', start_date, end_date, notional=10_000_000, leverage_ratio=5)
```

```python
def carry_trade_strat_forward_hedge(t_currency_rate, OIS_data, country_yields, zeros_data, forward_rate,
                      lending_currency, start_date, end_date, notional=10_000_000,
                        leverage_ratio=5, forward_hedge=False):
    """
    Implements a carry trade strategy for a given country and currency.

    Parameters:
    -----------
    t_currency_rate : pandas.DataFrame
        A DataFrame containing historical foreign exchange rates.
    OIS_data : pandas.DataFrame
        A DataFrame containing historical OIS rates.
    country_yields : pandas.DataFrame
        A DataFrame containing historical yields for the given country.
    zeros_data : pandas.DataFrame
        A DataFrame containing zero-coupon bond rates for the given country.
    lending_currency : str
        The currency in which the investor will lend.
    start_date : str or datetime
        The start date of the analysis (format: 'YYYY-MM-DD').
    end_date : str or datetime
        The end date of the analysis (format: 'YYYY-MM-DD').
    notional : float, optional (default=10_000_000)
        The initial notional amount to invest.
    leverage_ratio : float, optional (default=5)
        The leverage ratio for the carry trade strategy.

    Returns:
    --------

    float
        The cumulative profit or loss from the carry trade strategy.
    dict
        A dictionary of weekly profits or losses from the carry trade strategy.
    dict
        A dictionary of detailed profits or losses from the carry trade strategy.
    """
    zeros_data.columns = pd.to_datetime(zeros_data.columns)

    country_yields['tenor_mapped'] = country_yields['tenor_mapped'].astype(int)
    country_yields['date'] = pd.to_datetime(country_yields['date'])

    wkly_dates = get_wednesdays_with_adjustments(start_date, end_date, country_yields['date'].to_list())
    wkly_dates = wkly_dates.intersection(OIS_data['date'].to_list())
    wkly_dates = wkly_dates.intersection(zeros_data.columns.to_list())

    detailed_pnls = {}
    weekly_pnls = {}
    debt = notional * (1-(1/leverage_ratio))
    cumulative_pnl = 0

    for i, date in enumerate(wkly_dates):
        if i == len(wkly_dates) - 1:
            break
        if wkly_dates[i+1] - date > timedelta(days=8):
            continue


        row_index = OIS_data[OIS_data['date'] == date].index[0]
        OIS_rate = OIS_data.loc[row_index,'OIS']
        borrowing_rate = (OIS_rate + 0.005) / 100 / 52
        # borrowing_rate = (OIS_rate + 0.005) / 52

        # get lending rate
        lending_rate = country_yields[(country_yields['tenor_mapped'] == 5) & (country_yields['date'] == date)]['yield'].values[0]

        if lending_rate / 52  < borrowing_rate:
            continue
        else:
            #  t = 0
            initial_fx_rate_foreign_chf = t_currency_rate[(t_currency_rate['date'] == date) &
                                                  (t_currency_rate['code'] == lending_currency)]['rate'].values[0]
            initial_market_val = notional * initial_fx_rate_foreign_chf
            initial_price = bond_price(zeros_data[date], lending_rate, 5)
            qty_bonds = initial_market_val / initial_price

            #  t = 1
            next_date = wkly_dates[i+1]
            time_elapsed = (next_date - date).days / 365
            tenor = 5 - time_elapsed
            final_price = bond_price(zeros_data[next_date], lending_rate, tenor)
            final_market_val = final_price * qty_bonds
            bond_price_change = final_market_val - initial_market_val
            if forward_hedge:
                final_fx_rate_foreign_chf = forward_rate[(forward_rate.index == date)][lending_currency].values[0]
            else:
                final_fx_rate_foreign_chf = t_currency_rate[(t_currency_rate['date'] == next_date) &
                                                    (t_currency_rate['code'] == lending_currency)]['rate'].values[0]
            bond_price_change_chf = bond_price_change / final_fx_rate_foreign_chf

            #  t = 0
            initial_fx_rate_chf = t_currency_rate[(t_currency_rate['date'] == date) &
                                              (t_currency_rate['code'] == lending_currency)]['rate'].values[0]
            debt_lending = debt * initial_fx_rate_chf
            #  t = 1
            interest_payment = borrowing_rate * (debt_lending)
            if forward_hedge:
                final_fx_rate_chf = forward_rate[(forward_rate.index == date)][lending_currency].values[0]
            else:
                final_fx_rate_chf = t_currency_rate[(t_currency_rate['date'] == next_date) &
                                                (t_currency_rate['code'] == lending_currency)]['rate'].values[0]
            interest_payment_chf = interest_payment / final_fx_rate_chf
            fx_pnl_chf = (debt * initial_fx_rate_chf / final_fx_rate_chf) - debt

            total_pnl = bond_price_change_chf - interest_payment_chf + fx_pnl_chf
            cumulative_pnl += total_pnl

            weekly_pnls[date] = total_pnl
            # detailed_pnls[next_date] = {
            #                     'bond_price_change_usd': bond_price_change_chf,
            #                     'interest_payment_usd': interest_payment_chf,
            #                     'fx_pnl_usd': fx_pnl_chf,
            #                     'total_pnl': total_pnl}
            detailed_pnls[date] = {
                                'bond_price_change_usd': bond_price_change_chf,
                                'interest_payment_usd': interest_payment_chf,
                                'fx_pnl_usd': fx_pnl_chf,
                                'total_pnl': total_pnl}

    return cumulative_pnl, weekly_pnls, detailed_pnls
```

### **Forward Hedging Analysis**

Carry trade trading strategy contains significant foreign exchange (FX) risk due to fluctuations in exchange rates. To mitigate this risk, we use forward contracts to hedge against unfavorable currency movements, a technique known as forward hedging in a carry trade. To be specific, we locks in the future exchange rate using forward contracts with **Risk Neutral FX rate** instead of relying on the spot exchange rate at the settlement date.

```python
def tailMetrics(returns, quantile=.05, relative=False, mdd=True):
    metrics = pd.DataFrame(index=returns.columns)
    metrics['Skewness'] = returns.skew()
    metrics['Kurtosis'] = returns.kurtosis()

    VaR = returns.quantile(quantile)
    CVaR = (returns[returns < returns.quantile(quantile)]).mean()

    if relative:
        VaR = (VaR - returns.mean())/returns.std()
        CVaR = (CVaR - returns.mean())/returns.std()

    metrics[f'VaR ({quantile})'] = VaR
    metrics[f'CVaR ({quantile})'] = CVaR

    return metrics


def plotGraph(df, cur):
    # Plot
    plt.figure(figsize=(10, 5))
    plt.plot(df.index, df["Hedged PnL"], marker="o", label="Hedged PnL", linestyle="-")
    plt.plot(df.index, df["Unhedged PnL"], marker="s", label="Unhedged PnL", linestyle="--")
    plt.axhline(0, color="black", linewidth=0.8, linestyle=":")

    # Formatting
    plt.xlabel("Date")
    plt.ylabel("Profit and Loss (PnL)")
    plt.title(cur+" Weekly PnLs: Carry Trade Strategy (Hedged vs. Unhedged)")
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)

    # Show plot
    plt.show()

from datetime import timedelta

def analysis(wkly_pnls, wkly_pnls_forward, wkly_dates, close_prices, filtered_forward, lending_currency):
    max_datetime = max(wkly_dates)
    new_datetime = max_datetime + timedelta(weeks=1)

    current_index = pd.DatetimeIndex(wkly_dates)  # convert dict_keys to DatetimeIndex
    updated_wkly_dates = current_index.append(pd.DatetimeIndex([new_datetime]))

    # print(new_datetime, wkly_dates)

    # updated_wkly_dates = wkly_dates.append(pd.DatetimeIndex([new_datetime]))

    fx_rate_trades = close_prices[close_prices.index.isin(updated_wkly_dates)][[lending_currency]].copy(deep=True)
    forward_rate_trades = filtered_forward[filtered_forward.index.isin(updated_wkly_dates)][[lending_currency]].copy(deep=True)
    # fx_rate_trades = close_prices[close_prices.index.isin(wkly_pnls.keys())][[lending_currency]].copy(deep=True)
    # forward_rate_trades = filtered_forward[filtered_forward.index.isin(wkly_pnls.keys())][[lending_currency]].copy(deep=True)
    fx_trades_combined = pd.concat([fx_rate_trades, forward_rate_trades], axis=1)
    fx_trades_combined.columns = ['FX Rate Trades', 'Forward Rate Signing']
    fx_trades_combined['Forward Points'] = fx_trades_combined['Forward Rate Signing'] - fx_trades_combined['FX Rate Trades']
    fx_trades_combined['Forward Rate Executing'] = fx_trades_combined['Forward Rate Signing'].shift(1)
    fx_trades_combined['FX Rate Pct Change'] = fx_trades_combined["FX Rate Trades"].pct_change()
    fx_trades_combined['FX Rate Pct Change'] = fx_trades_combined['FX Rate Pct Change'].shift(-1)
    display(fx_trades_combined[:-1])

    lendingDf = pd.DataFrame({"Hedged PnL": wkly_pnls_forward.values(), "Unhedged PnL": wkly_pnls.values()}, index=wkly_pnls_forward.keys())

    plotGraph(lendingDf, lending_currency.split('_')[-1])
    display(lendingDf.describe())
    display(tailMetrics(lendingDf))
```

In some cases, hedging with forwards results in higher risk exposure and negative PnLs due to unfavorable forward rate adjustments. Additionally, hedging eliminates potential upside gains when FX rate movements favor the carry trade.

A more flexible approach is to use options contracts for risk management. By purchasing call or put options, traders pay a premium but retain the ability to profit when FX rate movements are favorable. If FX rates move against the carry trade, the option provides a safety net to hedge potential losses, while if the rates move in favor of the trade, the trader can let the option expire and still capture gains.

This strategy offers a more dynamic risk management solution, allowing traders to benefit from positive FX shifts while limiting downside risk, unlike forwards which lock in rates regardless of market movement.

## **Carry Trade with FX Option**




### **Hedging with USD/CHF Options**
**The Scenario Carry Trade Setup:**

We borrow in CHF (funding currency) at a low interest rate and invest in other currencies (like USD, EUR, GBP) that offer higher yields. Our profit depends on both the interest rate differential and stable or favorable FX movements.

**FX Risk:**

If CHF strengthens, the value of foreign-currency assets (when converted back to CHF) falls. For example, if we have USD assets and USD/CHF (CHF per USD) declines because CHF is strong, our CHF‐denominated returns are reduced.

To hedge against the risk of a strong CHF (i.e. a decline in the USD/CHF exchange rate), we can buy a **put option on USD/CHF**. This option gives us the right, but not the obligation, to sell USD at a predetermined rate (strike price, $K$). If CHF strengthens and the spot rate $S$ (CHF per USD) falls below $K$, our put option will be "in the money" and its payoff will help offset the losses on our carry trade.

**Underlying Logic:**

1. **Without the Hedge:** Suppose we have an asset with a notional value of $A$ USD. Its value in CHF at time $T$ is:

   $$\text{Value}_\text{unhedged} = A \times S_T$$
   
   where $S_T$ is the spot USD/CHF rate at maturity.

2. **With the Hedge (Buying a Put Option):** We buy a European put option on USD/CHF with strike $K$ and maturity $T$. The option's payoff per USD is:

   $$\text{Payoff}_\text{option} = \max(K - S_T, 0)$$
   
   If we hedge the entire $A$ USD exposure, our combined CHF value becomes:
   
   $$V = A \times S_T + A \times \max(K - S_T, 0)$$
   
   This can be written piecewise as:
   
   $$V = \begin{cases}
   A \times S_T, & \text{if } S_T \geq K, \\
   A \times K, & \text{if } S_T < K.
   \end{cases}$$

   **Interpretation:**
   * If USD/CHF stays above the strike $K$ (CHF is not overly strong), we simply convert at the prevailing market rate, $S_T$.
   * If USD/CHF falls below $K$ (CHF strengthens significantly), we exercise the option and effectively convert our USD at $K$, capping our downside.

3. **The Cost of Hedging – Option Premium:** The premium for the put option (using a version of the **Garman-Kohlhagen model for FX options**) is given by:

   $$P = K e^{-r_f T} N(-d_2) - S_0 e^{-r_d T} N(-d_1)$$
   
   where:
   * $S_0$ is the current USD/CHF spot rate,
   * $K$ is the strike,
   * $r_d$ and $r_f$ are the domestic and foreign risk-free rates (for FX options, these correspond to the interest rates in the respective currencies),
   * $\sigma$ is the volatility,
   * $T$ is the time to maturity,
   * $N(\cdot)$ is the cumulative distribution function of the standard normal distribution,
   * and $d_1 = \frac{\ln(S_0/K) + (r_f - r_d + \sigma^2/2)T}{\sigma \sqrt{T}}$, $d_2 = d_1 - \sigma \sqrt{T}$.
   
   This premium is a cost that reduces the net carry yield but provides protection against a strong CHF.

```python
def get_latest_sofr():
    try:
        # Get last 7 days (to account for weekends/holidays when rate doesn't update)
        end_date = datetime.now()
        start_date = end_date - timedelta(days=7)

        # SOFR series ID in FRED is 'SOFR'
        sofr_data = pdr.get_data_fred('SOFR', start=start_date.strftime('%Y-%m-%d'), end=end_date.strftime('%Y-%m-%d'))

        # Get the latest non-NaN value
        latest_sofr = sofr_data.dropna().iloc[-1, 0]
        latest_date = sofr_data.dropna().index[-1]

        return latest_sofr, latest_date
    except Exception as e:
        print(f"Error fetching SOFR data: {e}")
        return None, None

sofr_rate, sofr_date = get_latest_sofr()
sofr_rate, sofr_date
```

```python
def gk_put(S0, K, T, r_d, r_f, sigma):
    """
    Garman-Kohlhagen model for pricing a European put option on an FX rate.

    Parameters:
      S0    : Current spot rate (CHF per USD)
      K     : Strike price (CHF per USD)
      T     : Time to maturity (in years)
      r_d   : Domestic risk-free rate (CHF)
      r_f   : Foreign risk-free rate (USD)
      sigma : Annualized volatility

    Returns:
      put_price : The put option premium in CHF per USD.
    """
    d1 = (math.log(S0/K) + (r_d - r_f + 0.5 * sigma**2)*T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)
    put_price = K * math.exp(-r_d * T) * norm.cdf(-d2) - S0 * math.exp(-r_f * T) * norm.cdf(-d1)
    return put_price

# Parameters
A = 10_000_000                            # Notional amount in USD
r = usd_cum_pnl/A                           # Carry trade return (5% yield in USD terms)
S0 = usd_chf_data['4. close'].iloc[0]                # Current USD/CHF spot rate (CHF per USD)
K = S0 - 0.005                           # Strike of the put option (CHF per USD), 0.005 is threshold
T = 1/52                               # Time to maturity in years
r_d = 0.0044                             # Domestic risk-free rate (CHF)
r_f = sofr_rate/100                         # Foreign risk-free rate (USD)
sigma = 0.10                            # Annualized volatility

# Compute the put option premium using the Garman-Kohlhagen model
put_premium = gk_put(S0, K, T, r_d, r_f, sigma)
print(f"GK Put Option Premium: {put_premium:.4f} CHF per USD")

# Calculate the USD asset value after carry trade return
A_after = A * (1 + r)

# Simulate a range of future USD/CHF spot rates (S_T)
S_T = np.linspace(S0-0.05, S0+0.05, 100)

# Unhedged portfolio value in CHF: converting the USD asset at the future spot rate
unhedged_value = A_after * S_T

# Option payoff per USD: max(K - S_T, 0)
option_payoff = np.maximum(K - S_T, 0) * A_after

# Hedged portfolio value in CHF:
# Adding the option payoff and subtracting the total option premium cost (premium * total USD exposure)
hedged_value_option = A_after * S_T + option_payoff - put_premium * A_after

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(S_T, unhedged_value, label='Unhedged Value', color='blue')
plt.plot(S_T, hedged_value_option, label='Hedged Value (with GK Premium)', color='red', linestyle='--')
plt.axvline(x=K, color='gray', linestyle=':', label=f'Strike = {K}')
plt.xlabel('Future USD/CHF Spot Rate (CHF per USD)')
plt.ylabel('Portfolio Value in CHF')
plt.title('Hedging Carry Trade Exposure with USD/CHF Put Option (Garman-Kohlhagen Model)')
plt.legend()
plt.grid(True)
plt.show()

# Display sample results
for s, unhedged, hedged in zip(S_T[::10], unhedged_value[::10], hedged_value_option[::10]):
    print(f"Spot = {s:.3f}, Unhedged = {unhedged:,.0f} CHF, Hedged = {hedged:,.0f} CHF")
```

### **Option Hedging Analysis**

With T = 1week, we gain near‐term downside protection. Premiums are usually lower than for a multi‐month hedge, and are aligning with our carry trades expiry. $K$ = $S_0$ - 0.005 means we only hedge beyond a 0.005 drop in USD/CHF. This reduces premium cost but provides a partial hedge (a “knock‐in” effect at that level).0.005 is a threshold we get after analyzing the exchange rate volatility.

If CHF strengthens significantly below $K$, the put payoff can prevent large FX losses.
If USD/CHF stays above $K$, we lose only the premium.

Compare to Doing Nothing, the chart will show that once
$S_T$ dips below $K$, the hedged PnL (red line) is higher than the unhedged (blue line). For above $K$, the hedged line is slightly lower by the premium cost.

**Summary of the Hedge**

**Objective:** Protect the value of our USD (or other foreign currency) assets in CHF terms if CHF appreciates.
Hedge Instrument: A put option on USD/CHF with strike $K$.
Hedged Value:

If $S_T \geq K$: We get $A \times S_T$ CHF.

If $S_T < K$: We get $A \times K$ CHF.

**Trade-off:** The option premium $P$ is the cost of hedging, which reduces our net profit from the carry trade, but it limits downside risk if CHF becomes very strong.


**Direct vs. Indirect Hedge** *(Currently, we can only find available USD/CHF options from Exchange)*

**Direct Hedge:**

If we have exposure in USD (i.e., we lend in USD), a put option on USD/CHF directly protects us against a strong CHF (which means a lower USD/CHF rate).

**Indirect (Cross) Hedge:**

For currencies such as EUR or GBP, our risk is measured by the EUR/CHF or GBP/CHF rate. While these rates might be correlated with USD/CHF, the relationship isn't one-to-one. We could use a USD/CHF option to hedge some of the exposure if, for example, the USD acts as a proxy for overall CHF strength. However, this is an imperfect hedge because:

1. The correlations may not be stable.
The dynamics (volatility, skew, etc.) of EUR/CHF or GBP/CHF can differ from those of USD/CHF.

2. Underlying Logic and Mathematics for a Cross-Hedge
Let's say the carry trade has exposures in several currencies. The net CHF exposure from each asset is:
$V_i = A_i \times S_{i/CHF}$
where: $A_i$ is the notional amount in currency $i$.
$S_{i/CHF}$ is the spot exchange rate (CHF per unit of currency $i$).

If we hedge using USD/CHF options, we assume that movements in USD/CHF are correlated with movements in other cross rates. The effectiveness of our hedge will depend on the hedge ratio, which can be estimated using the correlation between the changes:

$\text{Hedge Ratio}{i} = \frac{\text{Cov}(\Delta S{USD/CHF}, \Delta S_{i/CHF})}{\text{Var}(\Delta S_{USD/CHF})}$

For example, if we have a EUR exposure, note that

$S_{EUR/CHF} \approx S_{EUR/USD} \times S_{USD/CHF}$

If $S_{EUR/USD}$ is relatively stable or we know its dynamics, we might adjust the USD/CHF hedge by this factor. In practice, we would determine a hedge ratio $\beta$ such that:
$\text{Hedge Amount (USD exposure)} = \beta \times \text{EUR Exposure (in CHF)}$
Then, using a USD/CHF put option with strike $K$, our hedged portfolio's CHF value would be:

$V = A_i \times S_{i/CHF} + \beta A_i \times \max(K - S_{USD/CHF}, 0)$

This hedge protects our overall CHF value if CHF strengthens, but because the USD/CHF and EUR/CHF (or GBP/CHF) moves aren't perfectly correlated, the hedge won't be exact.  


**Key Considerations**

1. Correlation Analysis: Need to analyze historical correlations between USD/CHF and the other currency pairs (EUR/CHF, GBP/CHF, etc.) to set the right hedge ratios. If the correlations are high, a USD/CHF option might serve as a reasonable proxy. Based on the correlation matrix below, think it is only approachable for HKD/CHF.

2. Volatility and Skew Differences: The implied volatility and skew of options on different currency pairs can differ. When hedging with a cross hedge, we might not fully capture the risk profile of the original exposure.

3. Dynamic Hedging: Because correlations can change over time, the hedge ratio may need to be adjusted frequently. This means that a static hedge using a USD/CHF option might only be partially effective.
